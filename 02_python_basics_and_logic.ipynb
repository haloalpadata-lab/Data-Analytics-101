## **LOOP**

```python
#FOR-LOOP
# Iterating through a range of numbers
for i in range(1, 6):
	print(f"Processing row number: {i}")

# Iterating through a list
fruits = ["Apple", "Banana", "Cherry"]
for fruit in fruits:
	print(f"Current fruit is: {fruit}")

#WHILE-LOOP
# System prompt that continues until user types ‘exit’
text = ""
while text.lower() != "exit":
	text = input("Enter a command (type ‘exit’ to quit):")
print(f"You typed: {text}")
```

## **DATA STRUCTURES**

```python
# LIST
## Creating and modifying a list
sales_data = [100, 200, 150, 300]
sales_data.append(250) # add a value at the end
sales_data.insert(0, 50) # add a value at zero index
sales_data.remove(150) # Remove a value
print(f"Updated sales data: {sales_data}")

# TUPLE
## Unpacking a tuple
coordinates = (10, 20)
x, y = coordinates
print(f"X-axis: {x}, Y-axis: {y}")

# SET
## Extracting unique items from a list
messy_ids = [1, 2, 2, 3, 3, 3, 4, 5]
unique_ids = set(messy_ids)
print(f"Unique Customer IDs: {unique_ids}")

# DICYIONARY
## Creating a customer profile dictionary
customer_profile = {
	"name": "John Doe",
	"age": 30,
	"status": "Active"
}

## Accessing and adding new data
print(customer_profile["name"]) # Output: John Doe
customer_profile["country"] = "USA" # Add new data

## Iterating through dictionary
for key, value in customer_profile.items():
  print(f"{key.capitalize()}: {value}")
```

## **FUNCTION**

```python
# Defining a function with parameters and return value
"""
This function calculates the net profit.
"""
def calculate_net_profit(revenue, expenses):
  net_profit = revenue - expenses
  return net_profit

# Calling the function
q1_profit = calculate_net_profit(5000, 2000)
print(f"Q1 Net Profit is: ${q1_profit}")

# Lambda function to add 10% tax
add_tax = lambda price: price * 1.10
print(add_tax(100)) # Output: 110.0
```

## **CASE STUDY**

### 1. Number Guessing Game


```python
import random

# 1. Setup the initial variables
secret_number = random.randint(1, 100)
attempts = 0
max_attempts = 7
print("Welcome! Guess the number between 1 and 100.")

while attempts < max_attempts:
  try :
    guess = int(input("Enter your guess : "))
    attempts +=1
  except ValueError:
    print("Invalid input. Please enter a valid integer.")
  else:
    if guess < secret_number:
      print("Too low! Try a higher number.")
    elif guess > secret_number:
      print("Too high! Try a lower number.")
    else:
      print(f"Congratulations! You guessed it in {attempts} attempts.")
      break;
# 3. Post-game check
if guess != secret_number:
  print(f"Game Over! The secret number was {secret_number}.")
```

### 2. Contact Book

```python
# 1. Initializing the dictionary (Inisialisasi kamus)
contact_book = {
	"John": "555-0101",
	"Alice": "555-0202",
	"Bob": "555-0303"
}

# 2. Searching logic
search_name = input("Enter the name you want to search for:").capitalize()

# 3. Validating key existence in dictionary
if search_name in contact_book:
	phone_number = contact_book[search_name]
	print(f"Contact Found! {search_name}’s number is {phone_number}.")
else:
	print(f"Error: {search_name} is not in the contact book.")
```
