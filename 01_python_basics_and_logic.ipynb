**User Input & Type Casting**

```python

num_str = input("Input the price : ")
num_float =  float(num_str)
tax = num_float * 0.11
print(f"Tax : {tax}")

# Input validation using try-except
try:
  n = int(input("Input the price"))
except:
  print("Invalid price input. Try again")
else:
  tax = n * 0.11
  print(f"Tax : {tax}")
```

**IF-ELSE-IF Structure**

```python

age = 20
id_card = True

if age >= 18 and id_card == True:
  print('Access is okay. You are allowed to create bank account!')
elif age >=18 and id_card == False:
  print('Access is denied. You must have an id_card first!')
else:
  print('Acces id denied!')

# Case Study
# 1. Password validation
passwd = ""
passwd = input("Input your password :")

if(len(passwd) < 8):
  print("Weak:Password must be at least 8 characters long")
elif "123" in passwd or "password" in passwd:
  print("Weak:Avoid using common pattern like '123' or 'password' ")
else:
  print("strong:Password is secure and accepted")

# 2. BMI (Body Mass Index) Calculator
# Ask user input and convert it to float
try:
  weight = float(input("Enter weight in kg:"))
  height = float(input("Enter height in meters:"))
except:
  print("Invalid input")
else:
  # BMI = weight / (height square)
  bmi = weight / (height ** 2)
  bmi_rounded = round(bmi, 2)
  print(f"\nYour BMI is: {bmi_rounded}")

  # Evaluation
  if bmi_rounded < 18.5:
    print("Category: Underweight")
  elif 18.5 <= bmi_rounded < 25:
    print("Category: Normal weight")
  elif 25 <= bmi_rounded < 30:
    print("Category: Overweight")
  else:
    print("Category: Obese")

```
