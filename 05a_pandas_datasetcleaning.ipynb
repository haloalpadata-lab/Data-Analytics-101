###**Clean Dataset**
Data cleaning means fixing bad data in your data set.
Bad data could be:
*   Empty cells
*   List item
*   Data in wrong format
*   Wrong data
*   Duplicates

## **CASE STUDY : CustomerCallList.csv**

```python
# Import panda as pd
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/DATASETS/cleanable-dataset/CustomerCallList.csv')

# 1. Drop all ghost column (resulted by converting xlxs file to csv file)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# 2. Clean Phone_Number data :
## Replace all of empty cell (NaN) to be empty string
df['Phone_Number'] = df['Phone_Number'].fillna('')

## Ensure all Phone_Number datas is converted to string
df['Phone_Number'] = df['Phone_Number'].astype(str)

## Clean  Phone_Number column using Regex
### Replace the unneeded characters except number with ''
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]', '', regex=True)

# 3. Clean 'Address', 'Paying Customer' and 'Do_Not_Contact' columns
## Create maping dictionary
mapping_dict = {'Y': 'Yes', 'N': 'No', 'N/a': '', 'nan':''}

## Clean Address column
df['Address'] = df['Address'].replace(mapping_dict)

## Clean 'Paying Customer column
df['Paying Customer'] = df['Paying Customer'].replace(mapping_dict)

## Clean 'Do_Not_Contact' column
df['Do_Not_Contact'] = df['Do_Not_Contact'].fillna('')
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace(mapping_dict)

# 4. Filtering Customer Call List
## Get only the paying customer and available to be contacted
df = df[(df['Paying Customer'] == 'Yes') & (df['Do_Not_Contact'] == 'No')]

## Get the only phone number is avilable
df = df[df['Phone_Number'] != '']

# 5. Reorder the index number
df = df.reset_index(drop=True)

# 6.  Remove 'Not_Useful_Column'
df = df.drop(columns=['Not_Useful_Column'])

# 7. Remove duplicate row
df = df.drop_duplicates()

# 8. Check the result
display(df)


```
